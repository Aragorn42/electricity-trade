```python
import pandas as pd
import numpy as np

# ==========================================
# 第一部分：生成 12列 x 5016行 的随机数据
# ==========================================
def generate_random_data(filename):
    n_samples = 5016
    n_models = 13

    # 1. 生成 12 个模型的随机预测 (0 或 1)
    # 模拟完全随机的“瞎猜”模型
    model_data = np.random.randint(0, 2, size=(n_samples, n_models))
    
    # 创建 DataFrame
    cols = [f"模型{i+1}预测结果" for i in range(n_models)]
    df = pd.DataFrame(model_data, columns=cols)

    # 2. 生成 Ground Truth (日前实时差价)
    # 为了让实验更有趣，我们在完全随机的基础上加一点点“投票倾向”
    # 这样旧投票法会有 >50% 的准确率，方便对比。
    # 如果完全随机，旧投票法就是50%。
    
    # 生成基础随机数 (-1 到 1)
    base_truth = np.random.uniform(-1, 1, size=n_samples)
    
    # (可选) 让真实结果稍微受模型投票影响 (模拟真实世界模型还是有点用的)
    # 如果你想要纯随机测试，可以注释掉下面这一行
    vote_influence = (model_data.sum(axis=1) - n_models/2) * 0.1 
    
    # 最终的差价
    df['日前实时差价'] = base_truth + vote_influence

    # 保存文件
    df.to_excel(filename, index=False)
    print(f"随机数据已生成: 12个模型, {n_samples}个样本 -> {filename}")

# ==========================================
# 第二部分：你的核心分析代码 (保持不变)
# ==========================================
def maximize_accuracy(file_path):
    # 1. 读取数据
    try:
        df = pd.read_excel(file_path)
    except Exception as e:
        print(f"读取文件失败: {e}")
        return

    # 动态获取模型列
    model_cols = [col for col in df.columns if "预测结果" in col]
    
    print(f"\n正在分析 {len(model_cols)} 个模型列...")

    ground_truth_col = '日前实时差价'
    # 转换为 0/1 标签
    df['true_label'] = df[ground_truth_col].apply(lambda x: 1 if x >= 0 else 0)
    
    for col in model_cols:
        df[col] = df[col].astype(int)

    # 拼接组合特征 (关键步骤)
    df['combo_pattern'] = df[model_cols].astype(str).agg(''.join, axis=1)

    # -------------------------------------------------------
    # 核心：计算查表逻辑
    # -------------------------------------------------------
    lookup_table = df.groupby('combo_pattern')['true_label'].agg(['count', 'mean']).reset_index()

    # 统计稀疏度 (看看有多少组合只出现过1次)
    single_count = len(lookup_table[lookup_table['count'] == 1])
    total_patterns = len(lookup_table)
    
    # 制定规则：胜率 > 0.5 判为 1，否则判为 0
    lookup_table['final_strategy'] = lookup_table['mean'].apply(lambda x: 1 if x > 0.5 else 0)

    # 回填策略
    df = pd.merge(df, lookup_table[['combo_pattern', 'final_strategy']], on='combo_pattern', how='left')

    # 计算新准确率
    df['is_correct'] = (df['final_strategy'] == df['true_label']).astype(int)
    new_accuracy = df['is_correct'].mean()

    # 计算旧投票法准确率 (过半数即为1)
    threshold = len(model_cols) / 2
    df['vote_sum'] = df[model_cols].sum(axis=1)
    df['old_vote_pred'] = df['vote_sum'].apply(lambda x: 1 if x > threshold else 0)
    old_accuracy = (df['old_vote_pred'] == df['true_label']).mean()

    # -------------------------------------------------------
    # 输出报告
    # -------------------------------------------------------
    print("\n" + "="*50)
    print("高维稀疏数据分析报告")
    print("="*50)
    print(f"样本总数: {len(df)}")
    print(f"模型数量: {len(model_cols)} (理论组合数: {2**len(model_cols)})")
    print(f"实际出现的组合数: {total_patterns}")
    print(f"只出现过1次的组合数: {single_count} (占比 {single_count/total_patterns:.1%})")
    print("-" * 50)
    print(f"旧方法(简单投票) 准确率: {old_accuracy:.2%}")
    print(f"新方法(查表策略) 准确率: {new_accuracy:.2%}")
    print(f"提升幅度: +{(new_accuracy - old_accuracy):.2%}")
    print("-" * 50)

    print("\n现象解释:")
    print("由于维度高(12个模型)且样本相对少，大量组合只出现了一次。")
    print("对于这些只出现一次的样本，查表法直接'记住'了正确答案，")
    print("因此你会看到极高的拟合准确率 (Overfitting)。")

    # 保存
    output_file = 'random_12models_result.xlsx'
    df.to_excel(output_file, index=False)
    print(f"\n结果已保存至: {output_file}")

# ==========================================
# 执行
# ==========================================
# 1. 生成数据
data_file = 'random_test_12_models.xlsx'
generate_random_data(data_file)

# 2. 运行模型
maximize_accuracy(data_file)
```
