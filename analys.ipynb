```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

file_path = 'output_report.xlsx'
gt_col = '日前实时差价'

def get_weights(file_path):
    print("训练并提取权重")
    df = pd.read_excel(file_path)
    
    y = (df[gt_col] > 0).astype(int)
    
    feature_cols = [c for c in df.columns if "准确率" in c]
    X_reconstructed = pd.DataFrame(index=df.index)

    for col in feature_cols:
        X_reconstructed[col] = np.where(df[col] == 1, y, 1 - y)

    model = LogisticRegression(fit_intercept=True, solver='liblinear')
    model.fit(X_reconstructed, y)
    
    weights = dict(zip(feature_cols, model.coef_[0]))
    bias = model.intercept_[0]
    
    print(f"提取到的常数项 (Bias): {bias:.4f}")
    print("提取到的权重 (Weights):")
    for k, v in weights.items():
        print(f"  {k}: {v:.4f}")

    return weights, bias, X_reconstructed, y

def verification(weights, bias, X_blind, y_true):
    df_inference = X_blind.copy()

    
    # z = w1*x1 + w2*x2 + ... + bias

    df_inference['final_score'] = bias
    
    # 逐列加上 (权重 * 预测值)
    for col_name, weight in weights.items():
        if col_name in df_inference.columns:
            df_inference['final_score'] += df_inference[col_name] * weight
    
    blind_predictions = (df_inference['final_score'] > 0).astype(int)
    final_acc = accuracy_score(y_true, blind_predictions)
    print(f"盲测验证准确率: {final_acc:.2%}")

weights, bias, X_clean, y_true = get_weights(file_path)
verification(weights, bias, X_clean, y_true)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import RFECV
from sklearn.model_selection import TimeSeriesSplit

# ================= 配置区域 =================
file_path = 'output_report.xlsx'
gt_col = '日前实时差价'
# ===========================================

def get_weights_auto_select(file_path):
    print("正在进行自动特征筛选与权重计算...")
    df = pd.read_excel(file_path)
    
    y = (df[gt_col] > 0).astype(int)
    
    # 1. 初始特征构建
    feature_cols = [c for c in df.columns if "准确率" in c]
    X_reconstructed = pd.DataFrame(index=df.index)
    for col in feature_cols:
        X_reconstructed[col] = np.where(df[col] == 1, y, 1 - y)
    
    # 2. 定义基础模型 (逻辑回归)
    # penalty='l1' 配合 solver='liblinear' 本身就有稀疏化(自动去噪)的效果，推荐加上
    base_model = LogisticRegression(fit_intercept=True, solver='liblinear', penalty='l2', C=1.0)
    
    # 3. 定义自动筛选器 (RFECV)
    # min_features_to_select: 最少保留几个模型 (例如至少保留3个)
    # cv: 使用时间序列交叉验证 (防止未来数据泄露)
    selector = RFECV(estimator=base_model, step=1, cv=TimeSeriesSplit(5), scoring='accuracy', min_features_to_select=20)
    
    print(f"开始在 {len(feature_cols)} 个模型中进行末位淘汰比赛...")
    selector.fit(X_reconstructed, y)
    
    # 4. 获取筛选结果
    selected_mask = selector.support_  # True/False 列表
    selected_features = [feature_cols[i] for i in range(len(feature_cols)) if selected_mask[i]]
    dropped_features = [feature_cols[i] for i in range(len(feature_cols)) if not selected_mask[i]]
    
    print(f"\n====== 筛选结果 ======")
    print(f"原始特征数: {len(feature_cols)}")
    print(f"保留特征数: {len(selected_features)}")
    print(f"被剔除的列 ({len(dropped_features)}个):")
    for f in dropped_features:
        print(f"  [x] {f}")
        
    # 5. 用最终筛选出的特征重新训练得到权重
    # 虽然 selector 已经fit过了，但为了提取直观的权重，我们对最优子集再跑一次
    X_final = X_reconstructed[selected_features]
    base_model.fit(X_final, y)
    
    weights = dict(zip(selected_features, base_model.coef_[0]))
    bias = base_model.intercept_[0]
    
    print(f"\n====== 最优权重 (仅包含保留列) ======")
    print(f"常数项 (Bias): {bias:.4f}")
    # 按权重绝对值排序输出
    sorted_weights = sorted(weights.items(), key=lambda x: abs(x[1]), reverse=True)
    for k, v in sorted_weights:
        print(f"  {k}: {v:.4f}")

    # 返回筛选后的X，以便盲测
    return weights, bias, X_final, y

def verification(weights, bias, X_blind, y_true):
    df_inference = X_blind.copy()
    df_inference['final_score'] = bias
    
    for col_name, weight in weights.items():
        if col_name in df_inference.columns:
            df_inference['final_score'] += df_inference[col_name] * weight
    
    blind_predictions = (df_inference['final_score'] > 0).astype(int)
    final_acc = accuracy_score(y_true, blind_predictions)
    print(f"\n盲测验证准确率 (最优子集): {final_acc:.2%}")

# 运行
weights, bias, X_clean, y_true = get_weights_auto_select(file_path)
verification(weights, bias, X_clean, y_true)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# ================= 配置 =================
file_path = 'output_report.xlsx'
gt_col = '日前实时差价'

# 在这里设置你想生成的随机列的数量
# 尝试设置两个数值运行对比：
# 1. 小一点的数：比如 10 (应该只有 50% 左右准确率)
# 2. 大一点的数：比如和你的行数一样多，或者行数的 80% (准确率会飙升)
num_random_features = 2000  
# =======================================

def run_random_noise_experiment(file_path, n_features):
    print(f"--- 开始实验: 使用 {n_features} 个纯随机噪音列进行拟合 ---")
    
    # 1. 读取真实的 Y (保持不变)
    df = pd.read_excel(file_path)
    y = (df[gt_col] > 0).astype(int)
    n_samples = len(y)
    
    print(f"数据总行数 (样本数): {n_samples}")
    
    # 2. 生成纯随机的 X (替换原来的 X_reconstructed)
    # 生成 0 和 1 的随机整数，形状为 (样本数, 特征数)
    # np.random.seed(42) # 如果想复现结果可以取消注释
    X_random_matrix = np.random.randint(0, 2, size=(n_samples, n_features))
    
    # 转为 DataFrame 方便后续处理，列名假装叫 "Random_0", "Random_1"...
    feature_cols = [f"Random_{i}" for i in range(n_features)]
    X_reconstructed = pd.DataFrame(X_random_matrix, columns=feature_cols, index=df.index)

    # 3. 强行拟合 (Fit)
    # C=1e5 代表几乎不进行正则化惩罚，让模型尽情地去过拟合
    model = LogisticRegression(fit_intercept=True, solver='liblinear', C=1e5)
    model.fit(X_reconstructed, y)
    
    # 提取权重
    weights = dict(zip(feature_cols, model.coef_[0]))
    bias = model.intercept_[0]
    
    print("拟合完成。")
    # print(f"提取到的常数项: {bias:.4f}") 
    # print(f"前5个随机列的权重: {[weights[f'Random_{i}'] for i in range(5)]}")

    return weights, bias, X_reconstructed, y

def verification(weights, bias, X_blind, y_true):
    df_inference = X_blind.copy()
    
    # z = w1*x1 + w2*x2 + ... + bias
    df_inference['final_score'] = bias
    
    # 逐列加上 (权重 * 预测值)
    for col_name, weight in weights.items():
        if col_name in df_inference.columns:
            df_inference['final_score'] += df_inference[col_name] * weight
    
    blind_predictions = (df_inference['final_score'] > 0).astype(int)
    final_acc = accuracy_score(y_true, blind_predictions)
    
    print(f"【验证结果】使用随机噪音的准确率: {final_acc:.2%}")

# 运行实验
weights, bias, X_clean, y_true = run_random_noise_experiment(file_path, num_random_features)
verification(weights, bias, X_clean, y_true)
```
